```python
import psycopg2 as ps
import pandas as pd
import glob
from tkinter import filedialog, Tk
from tkinter import *
import os
from io import StringIO
```

```python
listOfFiles = []
def get_filenames():
    Tk().withdraw()
    print("Initializing Dialogue... \nPlease select a file.")
    tk_filenames = filedialog.askdirectory()
    tempDir = tk_filenames
    return tempDir
choosen_dir = get_filenames()

os.chdir(choosen_dir)

for file in glob.glob("*.csv"):
    listOfFiles.append(file)
    
for file in listOfFiles:
    if file == 'knury.csv':
        knuryDF = pd.read_csv(file,sep=',')
    elif file == 'lochy.csv':
        lochyDF = pd.read_csv(file,sep=',')
    elif file == 'mag.csv':
        magDF = pd.read_csv(file,sep=',')
    elif file == 'mlode.csv':
        mlodeDF = pd.read_csv(file,sep=',')
    elif file == 'roz.csv':
        rozplodowaDF = pd.read_csv(file,sep=',')
        
print("Files loading done")
```

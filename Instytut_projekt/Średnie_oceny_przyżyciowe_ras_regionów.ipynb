```python
import pandas as pd
from tkinter import Tk
from tkinter.filedialog import askopenfilenames


path_list = []
Tk().withdraw()
filename = askopenfilenames()
for path in filename:
    path_list.append(path)

df = pd.concat(
    map(pd.read_csv,path_list), ignore_index=True)
```

```python
def mean_on_column(name_of_race, name_of_column):
    rslt_df = df.loc[df['KRM'] == name_of_race,[name_of_column]].mean()[0]
    
    return rslt_df

def mean_on_column_round(name_of_race, name_of_column):
    rslt_df = df.loc[df['KRM'] == name_of_race,[name_of_column]].mean()[0]
    rslt_df = round(rslt_df, 2)
    return rslt_df


def sum_on_column(name_of_race, name_of_column):
    rslt_df = df.loc[df['KRM'] == name_of_race,[name_of_column]].sum()[0]
    rslt_df = round(rslt_df, 2)
    return rslt_df
```

```python
df['KDOC'] = pd.to_datetime(df['KDOC'])
df['KDUR'] = pd.to_datetime(df['KDUR']) 
df['Wiek'] =  df['KDOC'] - df['KDUR'] 
df['Średnia grubość słoniny standar. (mm)'] = (df.KY3 + df.KY5)/2
list_of_race= list(df['KRM'].unique())
list_of_column = ['KP','Wiek','KY1','Średnia grubość słoniny standar. (mm)','OKO','T_PRZYR','MIESO_PROC']
list_of_city = list(df['KS'].unique())
print(list_of_city)
display(df)
df1 = df.groupby(['KRM', 'KS'])
print('-'*50)
df1.first()
```

```python
def create_final_df():
    final_list = {}
    
    for i in list_of_column:
        temp_list = []
        
        for j in list_of_race:
            
            if i == 'KP':
                temp_series = df['KRM'].where(df['KP'] == 1).dropna().value_counts()
                temp_list.append(temp_series[j])
            
            else:
                if i != "Wiek":
                    temp_list.append(round(mean_on_column(j,i),2))
                else:
                    temp_list.append(mean_on_column(j,i),)
                
            
            final_list[i] = temp_list
    final_list
    return pd.DataFrame.from_dict(final_list, orient="index", columns=list_of_race).T
```

```python
df2 = create_final_df()
df2 = df2.groupby(['KRM', 'KS'])
df2.first()
```

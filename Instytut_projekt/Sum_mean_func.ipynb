```python
import pandas as pd
from tkinter import Tk
from tkinter.filedialog import askopenfilenames


path_list = []
Tk().withdraw()
filename = askopenfilenames()
for path in filename:
    path_list.append(path)

df = pd.concat(
    map(pd.read_csv,path_list), ignore_index=True)

```

```python
def mean_on_column(name_of_city, name_of_column):
    rslt_df = df.loc[df['KE'] == name_of_city,[name_of_column]].mean()[0]
    rslt_df = round(rslt_df, 2)
    
    return rslt_df

def sum_on_column(name_of_city, name_of_column):
    rslt_df = df.loc[df['KE'] == name_of_city,[name_of_column]].sum()[0]
    rslt_df = round(rslt_df, 2)
    return rslt_df
```

```python
list_of_column = ["KY1","KY3","KY5","OKO","T_PRZYR","MIESO_PROC","T_INDEX"]
```

```python
def create_final_df(func):
    final_list = {}

    for i in list_of_column:
        temp_list = []
        for j in list_of_cities:
            temp_list.append(func(j,i))
            final_list[i] = temp_list
    final_list
    return pd.DataFrame.from_dict(final_list, orient="index", columns=list_of_cities)
```
